```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer


# Initialize VADER sentiment analyzer
analyzer = SentimentIntensityAnalyzer()



# Function to classify sentiment
def get_vader_sentiment(text):
    scores = analyzer.polarity_scores(text)
    compound_score = scores['compound']  # Get the compound score for overall sentiment

    # Assign sentiment labels based on compound score
    if compound_score >= 0.05:
        return '1'
    elif compound_score <= -0.05:
        return '-1'
    else:
        return '0'

# Apply the function to your DataFrame
df['category'] = df['simpleText'].apply(get_vader_sentiment)

# Save the DataFrame with sentiment labels
# df.to_csv('comments_with_sentiment.csv', index=False)


```
